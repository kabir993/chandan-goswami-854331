```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset['ApplicantIncome'].fillna(dataset['ApplicantIncome'].mean(),inplace=True)
```

```python
sns.distplot(dataset['ApplicantIncome'])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python

ss=StandardScaler()
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[['ApplicantIncome']]),columns=["x"])
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('before')
sns.distplot(dataset['ApplicantIncome_ss'])
plt.subplot(1,2,2)
plt.title('after')

sns.distplot(dataset['ApplicantIncome'])
plt.show()
#
```
