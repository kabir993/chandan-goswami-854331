```python
import pandas as pd
```

```python
dataset=pd.read_csv('/content/placement.csv')
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1].values
y=dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.score(x_train,y_train)
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
dt.score(x_train,y_train)
```

```python
sv=SVR()
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)
```

```python
sv.score(x_train,y_train)
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li=[("dt1",dt),("lr1",lr),("sv2",sv)]
```

```python
vr=VotingRegressor(estimators=li,weights=[5,25,30])
```

```python
vr.fit(x_train,y_train)
```

```python
vr.score(x_test,y_test)
```

```python
vr.score(x_train,y_train)
```

```python
prd={"dt":dt.predict(x_test),"lr":lr.predict(x_test),"sv":sv.predict(x_test),"vr":vr.predict(x_test)}
```
