```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/Student_Marks.csv")
dataset.head(20)
```

```python
dataset.corr()
```

```python
dataset.shape
```

```python
plt.scatter(dataset['time_study'],dataset['Marks'])
plt.xlabel('time_study')
plt.ylabel('Marks')
plt.show
```

```python
x=dataset[['time_study']]
y=dataset['Marks']
```

```python
from sklearn.preprocessing import  PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=3)
pf.fit(x)
x=pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
prD=lr.predict(x)
```

```python
plt.scatter(dataset['time_study'],dataset['Marks'])
plt.plot(dataset['time_study'],prD,color='red')
plt.title('Marks vs Study')
plt.xlabel('time_study')
plt.ylabel('Marks')

plt.show
```
