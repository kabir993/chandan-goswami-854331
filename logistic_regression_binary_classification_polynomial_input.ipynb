```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/content/Social_Network_Ads (1).csv")
dataset.head(4)
```

```python
dataset.drop(columns=["User ID","Gender"],inplace=True)
dataset.head(4)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Age",y="EstimatedSalary",data=dataset,hue="Purchased")
plt.show()
```

```python
x=dataset.iloc[:,:-1].values

```

```python
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# Since 'x' is already a NumPy array, you don't need to call 'to_numpy()' on it.
# Similarly, if 'y' is a Pandas Series, you can convert it to a NumPy array using 'y.values' instead of 'y.to_numpy()' if the to_numpy gives error.
plot_decision_regions(X=x, # Changed x.to_numpy() to x
                      y=y.values, # Changed y.to_numpy() to y.values if y is pandas Series else keep it y
                      clf=lr)
plt.show()
```
