```python
import pandas as pd
```

```python
dataset= pd.read_csv("/content/train_u6lujuX_CVtuZ9i (1).csv")
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset['Gender'].fillna(dataset['Gender'].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset['Married'].fillna(dataset['Married'].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
en_data=dataset[['Gender','Married']]
```

```python
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
Ohe=OneHotEncoder() # Define the OneHotEncoder object as Ohe
Ohe.fit_transform(en_data) # Use Ohe (capital 'O') to call the fit_transform method, maintaining consistency.
```

```python
ar=Ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar, columns=['Gender_Female', 'Gender_Male', 'Married_No', 'Married_Yes'])
```
