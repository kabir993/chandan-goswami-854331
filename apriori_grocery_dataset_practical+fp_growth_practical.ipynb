```python

from mlxtend.preprocessing import TransactionEncoder # This line correctly imports the class
tr = TransactionEncoder() # Use TransactionEncoder with an uppercase 'E' here
```

```python
tr.fit(Market)
```

```python
df = pd.DataFrame(tr.transform(Market),columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support=0.07, use_colnames=True, max_len=3)
frequent_itemsets.sort_values(by='support', ascending=False).head(20)
```

```python
from mlxtend.frequent_patterns import fpgrowth
fpgrowth(df, min_support=0.07, use_colnames=True, max_len=3).sort_values(by='support')
```
