```python
x=dataset.iloc[:,:-1].values
y=dataset['Purchased']
```

```python
from imblearn.under_sampling import RandomUnderSampler # Import from imblearn.under_sampling instead of imblearn.over_sampling
```

```python
ru=RandomUnderSampler()
x_ru,y_ru=ru.fit_resample(x,y)
```

```python
 y_ru.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_ru,y_ru,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
